```python
import matplotlib.pyplot as plt
import json
```

```python
f = open("../data/ca_gb/ca_gb_cleaned_ranked.json", "r")
rank = [entry["rank"] for entry in json.load(f)]
fig, ax = plt.subplots(tight_layout=True)
hist = ax.hist(rank, bins=range(10))
```
